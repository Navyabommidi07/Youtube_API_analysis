```python
# Creating module to reuse the functions 
from googleapiclient.discovery import build
import pandas as pd

def initialize_youtube(api_key):
    return build('youtube', 'v3', developerKey=api_key)

def get_channel_videos(youtube, channel_id):
    res = youtube.channels().list(id=channel_id, part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']

    videos = []
    next_page_token = None

    while True:
        res = youtube.playlistItems().list(playlistId=playlist_id, part='snippet', maxResults=50, pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')

        if next_page_token is None:
            break

    return videos

def extract_video_details(videos):
    video_details = []
    for video in videos:
        video_data = {
            'title': video['snippet']['title'],
            'videoId': video['snippet']['resourceId']['videoId'],
            'publishedAt': video['snippet']['publishedAt']
        }
        video_details.append(video_data)
    return video_details

def get_video_statistics(youtube, video_ids):
    stats = []
    for i in range(0, len(video_ids), 1050):  # API allows a maximum of 50 ids per request
        res = youtube.videos().list(id=','.join(video_ids[i:i+50]), part='statistics').execute()
        stats += res['items']
    return stats

def add_statistics_to_details(video_details, video_stats):
    for video_detail in video_details:
        video_id = video_detail['videoId']
        for stat in video_stats:
            if stat['id'] == video_id:
                video_detail['viewCount'] = int(stat['statistics'].get('viewCount', 0))
                video_detail['likeCount'] = int(stat['statistics'].get('likeCount', 0))
                video_detail['commentCount'] = int(stat['statistics'].get('commentCount', 0))
                break
    return video_details
```
